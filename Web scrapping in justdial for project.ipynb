```python
import numpy as np
final = pd.DataFrame()
    


for k in range(1,20):
    
    url = 'https://www.justdial.com/Vellore/Banquet-Halls/nct-10035861?page={}'.format(k)
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get( url , headers = headers).text
    
    soup = BeautifulSoup(webpage, 'lxml')
     
    category_name = []
    namee = []
    rating = []
    votess = []
    address = []
    
    
    
    for l in range(0, 10):
        category_name.append('Banquet_Halls') #category_name
        address.append(soup.find_all('a',class_ = 'morehvr')[l].text.strip()[12:]) #address
        votess.append((soup.find_all('span',class_ = 'lng_vote')[l].text)) #no of votes
        rating.append(soup.find_all('span',class_ = 'green-box')[l].text.strip())  #reviews
        namee.append(soup.find_all('span',class_ = 'lng_cont_name')[l].text.strip())  #max 9 values
        
    finans = []
    for j in votess:

        for i in j.split():
            if i.isdigit():
                finans.append(i) 
                
                
    d = {'category_name': category_name,  'name': namee, 'rating': rating,'votes':finans, 'address': address}
    df = pd.DataFrame(d)
    
    final = pd.concat([final,df], ignore_index = True)

    
        
    

        
     
  
    
```
